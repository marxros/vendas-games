```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
# lendo csv
videogames = pd.read_csv('vgsales.csv')
```

```python
videogames.head(10)
```

```python
videogames.describe()
```

```python
# Renomeando colunas
videogames.columns = ['Ranking', 'Nome', 'Plataforma', 'Ano', 'Gênero',
                     'Editora', 'Vendas América do Norte', 'Vendas EUA', 
                     'Vendas Japão', 'Outras Vendas', 'Vendas Global']
```

```python
videogames.head(10)
```

```python
# verificando onde não há ano de lançamento definido
videogames[videogames['Ano'].isnull()].head()
```

```python
# Jogos lançados por plataforma
jogos_lancados_por_plataforma = videogames['Plataforma'].value_counts()
jogos_lancados_por_plataforma.plot()
```

```python
# alterando para grafico de barras
jogos_lancados_por_plataforma.head(10).plot(kind = 'bar', figsize=(11,5), grid=False, rot=0, color='blue')

plt.title('Os 10 videogames com mais titulos lançados')
plt.xlabel('Videogame')
plt.ylabel('Quantidade de jogos lançados')
plt.show()
```

```python
top_10_vendidos = videogames[['Nome', 'Vendas Global']].head(10).set_index('Nome').sort_values('Vendas Global', ascending=True)
top_10_vendidos.plot(kind='barh', figsize=(11,7), grid=False, color='darkred', legend=False)

plt.title('Os 10 jogos mais vendidos no mundo')
plt.xlabel('Total de vendas (em milhões de doláres)')
plt.show()
```

```python
# criando um mapa de calor
# a função do crosstab do panda, cruza os dados
crosstab_vg = pd.crosstab(videogames['Plataforma'], videogames['Gênero'])
crosstab_vg.head()
```

```python
# adiconando a coluna Total no dataframe
crosstab_vg['Total'] = crosstab_vg.sum(axis=1)
crosstab_vg.head()
```

```python
# agrupando os 10 consoles que mais tivem lançamento e que possuem mais de 1000 jogos lançados
top10_plataforms = crosstab_vg[crosstab_vg['Total']>1000].sort_values('Total', ascending = False)

# pega o dataframe anterior e adiciona a coluna total no final
top10_final = top10_plataforms.append(pd.DataFrame(top10_plataforms.sum(), columns=['total']).T, ignore_index=False)

# definindo tamanho da fonte
sns.set(font_scale=1)

# definindo tamanho do gráfico
plt.figure(figsize=(18,9))

# heatmap do seaborn para personalizar o gráfico
# annot para aparecer os numeros dentro dos quadrados
# vmax e vmin é para definir os valores máximo e mínimo,quanto mais próximo ao max mais escuro a célula.
sns.heatmap(top10_final, annot=True, vmax=top10_final.loc[:'PS', :'Strategy'].values.max(), vmin=top10_final.loc[:, :'Strategy'].values.min(), fmt='d')
plt.title('QUANTIDADE DE TÍTULOS POR GÊNERO E CONSOLE')
plt.xlabel('GÊNERO')
plt.ylabel('CONSOLE')
plt.show()
```
